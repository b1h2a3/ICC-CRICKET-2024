```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("D:/ICC Mens T20 Worldcup.csv",encoding= 'unicode_escape')
```

```python
ax = sns.countplot(x='Highest Score',data=df)
sns.set(rc={'figure.figsize':(15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
highest_score = df.groupby(['Highest Score'],as_index=False)['Top Scorer'].sum().sort_values(by='Top Scorer',ascending=False)
sns.barplot(x='Highest Score',y='Top Scorer',data=highest_score)
sns.set(rc={'figure.figsize':(15,10)})
```

```python
df.columns
```

```python
ax = sns.countplot(x='Winners',data=df)
sns.set(rc={'figure.figsize':(15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
winner = df.groupby(['Winners'],as_index=False)['Won by'].sum().sort_values(by='Won by',ascending=False)
sns.barplot(x='Winners',y='Won by',data=winner)
sns.set(rc={'figure.figsize':(35,20)})
```

```python
ax = sns.countplot(x='Player Of The Match',data=df)
sns.set(rc={'figure.figsize':(15,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
MVP = df.groupby(['Winners'],as_index=False)['Player Of The Match'].sum()
sns.barplot(x='Winners',y='Player Of The Match',data=MVP)
```
